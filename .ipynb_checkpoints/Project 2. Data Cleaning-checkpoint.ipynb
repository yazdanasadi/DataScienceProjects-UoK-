```python
df_cols2=['acoustic_neuroma',
         'bells_palsy',
         'cochlear_age' ,
         'cochlear_age_and_noise',
         'cochlear_age_plus_poss_menieres',
         'cochlear_noise_and_heredity',
         'cochlear_poss_noise',
         'cochlear_unknown',
         'conductive_discontinuity',
         'conductive_fixation',
         'mixed_cochlear_age_fixation',
         'mixed_cochlear_age_otitis_media',
          'mixed_cochlear_age_s_om',
          'mixed_cochlear_unk_discontinuity',
          'mixed_cochlear_unk_fixation',
          'mixed_cochlear_unk_ser_om',
          'mixed_poss_central_om',
          'mixed_poss_noise_om',
          'normal_ear',
          'otitis_media',
          'poss_central',
          'possible_brainstem_disorder',
         'possible_menieres',
         'retrocochlear_unknown',
          ]

df2 = pd.read_csv("./audiology.data",
                  header=None,
                  names=df_cols2,
                  na_values=['?',' ?'])

df2.sample(5)
```

```python
ds = 'aduiology Database' 
cn = 'normal_ear'

print(f'{"data set name:":<20} {ds}')
print(f'{"calss name:":<20} {cn}' + '(calssification feature)')
```

```python
# ---------------------- Define a Function for describe any Dataset  --------
def describe(df, dataset_name, class_name):
    
    print(dataset_name+"\n")
    
    print("Number of Instances:" ,df.shape[0])
    print()
    
    print("Number of Attributes:", df.shape[1])
    print()
    
    print("Attributes Information:")
    print(df.dtypes)
    
    print("\nClass Distribution:")
    print( df[class_name].value_counts())
```

```python
#MISSING VALUES
print('Any missing value?', df2.isnull().values.any())
print('Number of Missing values', df2.isnull().sum())
```

```python
#drop missing value 
if df2.isnull().values.any():
    
    deleted = df2.dropna()
    
    data_loss = ( len(df2) - len(deleted) ) / len(df2)
    
    if data_loss < .4 : # if data loss was lower then 40% delete the NaN rows
        deleted.reset_index(drop=True, inplace=True)
        df2 = deleted
        print('nan values deleted', ds)
        
    else:
        logging.warning("can't delete null values from: "+ds
                        +f' ({data_loss*100:.2f} % data loss!)')
        
print('Any missing value?', df2.isnull().values.any()) # for test
```

# Adult Part 2

```python
df_cols=['age',
         'workclass',
         'fnlwgt' ,
         'education',
         'education-num',
         'marital-status',
         'occupation',
         'relationship',
         'race',
         'sex',
         'capital-gain',
         'capital-loss',
         'hours-per-week',
         'native-country',
         'income']

df = pd.read_csv("./adult.data",
                  header=None,
                  names=df_cols,
                  na_values=['?',' ?'])

df.sample(5)
```

```python
# ---------------------- Define a Function for describe any Dataset  --------
def describe(df, dataset_name, class_name):
    
    print(dataset_name+"\n")
    
    print("Number of Instances:" ,df.shape[0])
    print()
    
    print("Number of Attributes:", df.shape[1])
    print()
    
    print("Attributes Information:")
    print(df.dtypes)
    
    print("\nClass Distribution:")
    print( df[class_name].value_counts())
```

```python
ds_name = 'Adult Database' 
class_name = 'income'

print(f'{"data set name:":<20} {ds_name}')
print(f'{"calss name:":<20} {class_name}' + '(calssification feature)')
```

```python
describe(df, ds_name, class_name)
```

```python
#MISSING VALUES
print('Any missing value?', df.isnull().values.any())
print('Number of Missing values', df.isnull().sum())
```

```python
#drop missing value 
if df.isnull().values.any():
    
    deleted = df.dropna()
    
    data_loss = ( len(df) - len(deleted) ) / len(df)
    
    if data_loss < .1 : # if data loss was lower then 10% delete the NaN rows
        deleted.reset_index(drop=True, inplace=True)
        df = deleted
        print('nan values deleted', ds_name)
        
    else:
        logging.warning("can't delete null values from: "+ds_name
                        +f' ({data_loss*100:.2f} % data loss!)')
        
print('Any missing value?', df.isnull().values.any()) # for test
```

```python
# duplicated value
print('Any duplicated value?', df.duplicated().any())

# drop duplicate value
if df.duplicated().any():
    
    deleted = df.drop_duplicates()
    
    data_loss = ( len(df) - len(deleted) ) / len(df)    # data loss
    
    if data_loss < .1 : # if data loss was lower then 10% delete the NaN rows
        deleted.reset_index(drop=True, inplace=True)
        df = deleted
        print('duplicated values droped from', ds_name)
        
    else:
        logging.warning("can't delete duplicated values from: "+ ds_name
                        +f' ({data_loss*100:.2f} % data loss!)')

print('Any duplicated value?', df.duplicated().any())
```
